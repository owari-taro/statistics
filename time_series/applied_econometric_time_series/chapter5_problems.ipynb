```r
y<-rep(0,101)
z<-rep(0,101)
ex<-c(0,1,rep(0,99))
ez<-rep(0,101)
for (i in 1:100){
    y[i+1]<-0.8*y[i]+0.2*z[i]+ex[i+1]
    z[i+1]<-0.2*y[i]+0.8*z[i]+ez[i+1]    
}
plot(y)
```
