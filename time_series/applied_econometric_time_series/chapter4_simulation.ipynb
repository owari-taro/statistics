## MA with unit roote
Example below show that MA(1) with unit-root has siginificant acf value at higer lag,which is not seen in stationary ma ts

```r
create_ma1<-function(b1){
    #b1 is a coef of ma(1)
    y<-rep(0,200)
    e<-rnorm(200)
    y[1]<-e[1]
    for (i in 1:199){
        y[i+1]<-e[i+1]-b1*e[i]
    }   
    return (y)
}
y1<-create_ma1(1)
acf(y1,lag.max=50)
```

```r
y1<-create_ma1(0.8)
acf(y1,lag.max=50)
```

## AR(1) with unit-root

```r
create_ar1<-function(a1){
    #b1 is a coef of ma(1)
    y<-rep(0,200)
    e<-rnorm(200)
    y[1]<-0
    for (i in 1:199){
        y[i+1]<-a1*y[i]+e[i+1]
    }   
    return (y)
}
y2<-create_ar1(1)
acf(y2,lag.max=50)
pacf(y2,lag.max=50)
```

## power of unit-root test(p231)

```r
library(tseries)
res<-rep(0,2000)
for(j in 1:2000){
    y<-rep(0,50)
    e<-rnorm(50)
    a0<-2
    a1<-0.9
    y[1]<-a0+e[1]

    for(i in 1:49){
        y[i+1]<-a0+a1*y[i]+e[i]  
    }
    print(y)
    #dickey fulller test
    test<-adf.test(y,k=0)
    test
    res[j]<-test$p.value
}
hist(res)
summary(res)
```
