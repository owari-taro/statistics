```r
#aaaaaaaaadaaa
z1<-c(0,1,rep(0,99))
#constant effect for first ten periods
z2<-c(0,rep(1,10),rep(0,90))
#gradual decrease
z3<-c(0,1,0.75,0.5,0.25,rep(0,96))
#white noise
e<-rnorm(101)

#
model1<-function(a0,a1,z,e){
    #ar1 with intervention
    y<-rep(0,101)
    for (i in 1:100){        
        y[i+1]<-a0+a1*y[i]+z[i]+e[i]
    }
    return (y)
}
a0<-0
a1<-0.5   
ts.plot(model1(a0,a1,z1,e))
ts.plot(model1(a0,a1,z2,e))
ts.plot(model1(a0,a1,z3,e))

```

```r
a0<-0
a1<--0.5
ts.plot(model1(a0,a1,z1,e))
ts.plot(model1(a0,a1,z2,e))
ts.plot(model1(a0,a1,z3,e))
```
